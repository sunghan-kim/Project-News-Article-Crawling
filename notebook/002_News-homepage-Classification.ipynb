## 2.2 기사 링크 중복 확인

- rawdata에 같은 기사가 서로 다른 폴더에 있는 경우가 있다.
- 그렇기 때문에 파일 경로(`Path`) 정보를 제외한 나머지 정보가 중복된 데이터들이 존재한다.
- 이 중복된 데이터들을 확인해보자.

```python
df = df_origin.drop(columns=["Path"], axis=1)
df.head()
```

```python
# 중복 확인
df[df.duplicated()].head(10)
```

```python
df[df.duplicated()].to_csv("duplicated.csv", index=False)
```

```python
# 중복 제거
df.drop_duplicates()
```

```python
len(df.drop_duplicates(['URL'], keep='first'))
```
