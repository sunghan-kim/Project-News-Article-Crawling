## 2.1 데이터프레임 읽기

- 이전에 생성한 기사 링크가 포함된 데이터프레임을 읽어온다.

```python
import pandas as pd

df_origin = pd.read_csv('news_info.csv')
df_origin.head()
```

```python
print(len(df_origin))
```

<br>

## 2.2 기사 링크 중복 확인

- rawdata에 같은 기사가 서로 다른 폴더에 있는 경우가 있다.
- 그렇기 때문에 파일 경로(`Path`) 정보를 제외한 나머지 정보가 중복된 데이터들이 존재한다.
- 이 중복된 데이터들을 확인해보자.

```python
df = df_origin.drop(columns=["Path"], axis=1)
df.head()
```

```python
# 중복 확인
df[df.duplicated()].head(10)
```

```python
# 중복 제거
df = df.drop_duplicates()
df.head(10)
```

<br>

## 2.3 URL에서 도메인 추출

```python
url_list = df["URL"].tolist()
homepage_list = [url.split("/")[2] for url in url_list]
homepage_list[:10]
```

<br>

## 2.4 도메인 주소를 포함하는 데이터프레임 생성

```python
homepage_df = pd.DataFrame({"Domain": homepage_list})
homepage_df.head()
```

<br>

## 2.5 도메인 주소별 개수 확인

```python
homepage_df["Domain"].value_counts()
```

## 2.7 도메인별 개수 정보 데이터프레임 생성

```python
domain_cnt_df = homepage_df["Domain"].value_counts().reset_index().rename(columns={"Domain": "Cnt", "index": "Domain"})
domain_cnt_df.head(10)
```

<br>

## 2.8 중복 제거된 홈페이지 정보 데이터프레임 생성

```python
unique_homepage_df = homepage_df.drop_duplicates().reset_index().drop(columns=["index"])
unique_homepage_df.head(10)
```

<br>

## 2.9 중복 제거된 홈페이지 정보와 도메인 개수 정보 결합

```python
result_df = unique_homepage_df.join(domain_cnt_df.set_index('Domain'), on='Domain')
result_df = result_df.sort_values(by=['Cnt'], ascending=False)
result_df = result_df.reset_index().drop(columns=['index'])
result_df.head(10)
```

<br>

## 2.8 홈페이지 정보 csv 파일 저장

```python
try:
    result_df.to_csv("homepage.csv", index=False)
    print("Saved")
except:
    print("Save Error")
```
